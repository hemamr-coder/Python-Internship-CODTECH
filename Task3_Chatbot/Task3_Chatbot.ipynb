```python
while True:
    user = input("You: ").lower()

    if "hello" in user:
        print("Bot: Hello! How can I help you?")

    elif "name" in user:
        print("Bot: I am your internship chatbot.")

    elif "python" in user:
        print("Bot: Python is a programming language.")

    elif "bye" in user:
        print("Bot: Goodbye!")
        break

    else:
        print("Bot: Sorry, I don't understand.")
```
